````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
````

````python
train = pd.read_csv('./_data/train_imputation.csv', index_col=0)#id를 인덱스로
test = pd.read_csv('./_data/test_imputation.csv', index_col=0)
````

````python
y_train = train.pop('target')
y_train = y_train.to_numpy()
````

````python
categorical_features = ['sex', 'cp', 'slope', 'ca','exang', 'restecg', 'thal']
numeric_features = ['age', 'trestbps', 'thalach', 'fbs','oldpeak', 'chol']
````

```
ignore_features = ['fbs','oldpeak', 'chol', 'restecg','trestbps', 'thalach']

```

````python
input_features = categorical_features+numeric_features
````

````python
x_train = train.loc[:,input_features].to_numpy()
````

````python
x_test = test.loc[:,input_features].to_numpy()
````

````python
import tensorflow as tf
import autokeras as ak
ak_model = ak.StructuredDataClassifier(
    overwrite=True, max_trials=10, loss='binary_crossentropy',seed=32
)
ak_model.fit(x_train, y_train, epochs=100, validation_split=0.3)
````

````python
best_model = ak_model.export_model()
best_model.summary()
````

````python
predictions = best_model.predict(x_test)
predictions
````

````python
result = pd.DataFrame({0})
````

````python
result['target'] =0
````

````python
for i, pred in  enumerate(predictions):
    if pred < 0.5:
        result.loc[i,'target']= 0
    else:
        result.loc[i,'target']= 1
result
````

````python
input_features = categorical_features+numeric_features
````

````python
c_train = train.loc[:,categorical_features]
n_train = train.loc[:,numeric_features]
c_test = test.loc[:,categorical_features]
n_test = test.loc[:,numeric_features]
````

````python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
````

````python
rbs = RobustScaler()
n_train_robust = rbs.fit_transform(n_train)
n_test_robust = rbs.transform(n_test)
````

````python
#robust
x_train = np.concatenate([c_train.to_numpy(), n_train_robust], axis=1)
#std
#x_train = np.concatenate([c_train.to_numpy(), n_train_stdsc], axis=1)
````

````python
x_test = np.concatenate([c_test.to_numpy(), n_test_robust], axis=1)
````

````python
import tensorflow as tf
import autokeras as ak
ak_model = ak.StructuredDataClassifier(
    overwrite=True, max_trials=10, loss='binary_crossentropy',seed=32
)
ak_model.fit(x_train, y_train, epochs=200, validation_split=0.3)
````

````python
best_model = ak_model.export_model()
best_model.summary()
````

````python
predictions = best_model.predict(x_test)
predictions
````

````python
submission = pd.read_csv('./_data/sample_submission.csv')
````

````python
for i, pred in  enumerate(predictions):
    if pred <= 0.5:
        submission.loc[i,'target']= 0
    else:
        submission.loc[i,'target']= 1
submission
````
